```python
# Dependencies
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.info()
```

```python
# print total players using unique player count in a dataframe
data = {"Total Players": [purchase_data['SN'].nunique()],}
player_count = pd.DataFrame(data,columns = ["Total Players"])

player_count.head()
```

```python
# create a summary dictionary
data = {"Number of Unique items": [purchase_data['Item ID'].nunique()],
        "Average Price" : [purchase_data['Price'].mean()],
        'Number of Purchases' : [purchase_data['Purchase ID'].count()],
        'Total Revenue' : [purchase_data['Price'].sum()],}
# convert summary dictionary into a dataframe 
purchasing_analysis = pd.DataFrame(data,columns = ["Number of Unique items",'Average Price','Number of Purchases','Total Revenue'])
# format $ values
purchasing_analysis['Average Price'] = purchasing_analysis['Average Price'].map('${:,.2f}'.format)
purchasing_analysis['Total Revenue'] = purchasing_analysis['Total Revenue'].map('${:,.2f}'.format)

purchasing_analysis.head()
```

```python
# summary dataframe by gender
demographic_analysis = purchase_data.groupby("Gender")['SN'].nunique().reset_index()
# set index to gender
demographic_analysis = demographic_analysis.set_index('Gender')
# change column name
demographic_analysis.columns = ['Total Count'] 
# calculate & format percentage 
demographic_analysis['Percentage of Players'] = 100* demographic_analysis['Total Count'] / demographic_analysis['Total Count'].sum()
demographic_analysis['Percentage of Players'] = demographic_analysis['Percentage of Players'].map('{:.2f}%'.format)

demographic_analysis.head()
```

```python
# dcitionary for aggregations
aggregation = {'Purchase ID':'count','Price':['mean','sum'],'SN':'nunique'}
# Summary dataframe by gender
purchase_analysis_gender = purchase_data.groupby("Gender").agg(aggregation).reset_index()
# set index to gender
purchase_analysis_gender = purchase_analysis_gender.set_index('Gender')
# change column names
purchase_analysis_gender.columns = ['Purchase Count','Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']
# Update values in Avg total purchase per person using calculation
purchase_analysis_gender['Avg Total Purchase per Person'] = purchase_analysis_gender['Total Purchase Value']/purchase_analysis_gender['Avg Total Purchase per Person']
# format $ values
purchase_analysis_gender['Average Purchase Price'] = purchase_analysis_gender['Average Purchase Price'].map('${:,.2f}'.format)
purchase_analysis_gender['Total Purchase Value'] = purchase_analysis_gender['Total Purchase Value'].map('${:,.2f}'.format)
purchase_analysis_gender['Avg Total Purchase per Person'] = purchase_analysis_gender['Avg Total Purchase per Person'].map('${:,.2f}'.format)

purchase_analysis_gender.head()
```

```python
# find out max age
max_age = purchase_data['Age'].max()
# create custom age bins
custom_bins = [0,9,14,19,24,29,34,39,max_age]
# create custom age labels
custom_labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
# summary dataframe by custom bins and calculate unique no of players in each bin
age_demographics = purchase_data.groupby(pd.cut(purchase_data.Age,bins=custom_bins,labels=custom_labels)).agg({'SN':'nunique'}).reset_index()
# set index to Age
age_demographics = age_demographics.set_index('Age')
# change column name
age_demographics.columns = ['Total Count']
# calculate percentage of players in each bin and format
age_demographics['Percentage of Players'] = 100* age_demographics['Total Count'] / age_demographics['Total Count'].sum()
age_demographics['Percentage of Players'] = age_demographics['Percentage of Players'].map('{:.2f}%'.format)

age_demographics.head(10)
```

```python
# dictionary for aggregations
aggregation = {'Purchase ID':'count','Price':['mean','sum']}
top_spenders = purchase_data.groupby('SN').agg(aggregation).reset_index()
# set index to player
top_spenders = top_spenders.set_index('SN')
# change column name
top_spenders.columns = ['Purchase Count','Average Purchase Price','Total Purchase Value']
# sort values 
top_spenders.sort_values(by='Total Purchase Value', inplace=True, ascending=False)
# format $ values
top_spenders['Average Purchase Price'] = top_spenders['Average Purchase Price'].map('${:,.2f}'.format)
top_spenders['Total Purchase Value'] = top_spenders['Total Purchase Value'].map('${:,.2f}'.format)

top_spenders.head()
```

```python
# dictionary for aggregations
aggregation = {'Purchase ID':'count','Price':['mean','sum']}
popular_items = purchase_data.groupby(['Item ID','Item Name']).agg(aggregation).reset_index()
# set index to player
popular_items = popular_items.set_index(['Item ID','Item Name'])
# change column name
popular_items.columns = ['Purchase Count','Average Purchase Price','Total Purchase Value']
# sort values 
popular_items.sort_values(by='Purchase Count', inplace=True, ascending=False)
# format $ values
popular_items['Average Purchase Price'] = popular_items['Average Purchase Price'].map('${:,.2f}'.format)
popular_items['Total Purchase Value'] = popular_items['Total Purchase Value'].map('${:,.2f}'.format)

popular_items.head()
```
